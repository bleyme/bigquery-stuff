## Basic Bigquery stuff

```python
# BigQuery Stuff
# Set basic info

import pandas as pd 
from google.colab import auth
auth.authenticate_user()

project_id = 'YOUR_PROJECT_ID'

%load_ext google.colab.data_table
```

## Rows into single column

```python
getRowsASColumns = pd.io.gbq.read_gbq(f'''

    SELECT date, STRING_AGG(CAST(visitId as STRING), ", ") as allvisitId
    FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801` 
    GROUP BY date

''', project_id=project_id)
```

## Date Manipulation

```python
DateManipulation = pd.io.gbq.read_gbq(f'''

  SELECT 
  CURRENT_DATE() AS CURRENT_DATE,
  DATE_TRUNC(CURRENT_DATE(), MONTH) AS FIRST_DAY_CURRENT_MONTH,
  DATE_SUB(DATE_TRUNC(DATE_ADD(CURRENT_DATE(), INTERVAL 1 MONTH), MONTH), INTERVAL 1 DAY) AS LAST_DAY_CURRENT_MONTH,
  DATE_SUB(DATE_TRUNC(CURRENT_DATE(), MONTH), INTERVAL 1 DAY) AS LAST_DAY_PREVIOUS_MONTH,
  DATE_TRUNC(DATE_ADD(CURRENT_DATE(), INTERVAL 1 MONTH), MONTH) AS FIRST_DAY_NEXT_MONTH

''', project_id=project_id)
```
